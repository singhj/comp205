```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
x = [int(x) for x in '28 26 42 29 16 21 25 35 30 36 37 41 20 26 38 26 10 18 10 31'.split()]
```

```python
y = [int(y) for y in '15 14 15 12 37 30 7 14 28 13 5 13 24 8 13 17 27 29 8 5'.split()]
import pandas as pd
```

```python
pd.DataFrame([x]).rename(index={0: "x"})
```

```python
xy = pd.concat([pd.DataFrame([x]).rename(index={0: "x"}), pd.DataFrame([y]).rename(index={0: "y"})]).transpose()
xy
sns.regplot(x='x', y='y',data=xy)
```
