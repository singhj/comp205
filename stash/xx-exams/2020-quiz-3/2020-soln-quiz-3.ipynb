```python
nyc_weather = pd.read_csv('NYC_weather.csv')
```

```python
dtypes = {"DATE": 'str', 'SNOW': 'float', 'SNWD': 'float'}
parse_dates = ['DATE']
usecols=["DATE", "SNOW", "SNWD"]
wthr = pd.read_csv('NYC_weather.csv', dtype=dtypes, parse_dates=parse_dates, usecols=usecols)
wthr
```

```python
def day_of_year(dt):
    return dt.timetuple().tm_yday
wthr['days'] = wthr.loc[:,['DATE']].applymap(day_of_year)
wthr2 = wthr.loc[wthr.loc[:,'days']<=90]
wthr2
```

```python
plt.figure(figsize=(15, 8)) 
noprint = plt.plot(wthr2.loc[:,'days'], wthr2.loc[:,['SNOW', 'SNWD']])
```

### 6. Quasar Intensity

The US Naval Patrol's radio signals locked into a distant object and recorded the readings captured in quasar.csv.
It turned out not to be from a celestial object at all!

Fit a 7th-degree polynomial model to the data. Can you identify the curve?

### <span style="color:blue">6. Rubric and Solution.[20 points]</span>

<span style="color:blue">

1. Setting up the model, **5 points**
    
1. Training **5 points**

1. Prediction **5 points**

1. Identifying the curve as a sinusoid **5 points**

</span>


```python
quasar = pd.read_csv('quasar.csv')
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
poly_model = make_pipeline(PolynomialFeatures(7),
                           LinearRegression())

xfit = np.linspace(0, 10, 1000)
xx = quasar['x']
yy = quasar['y']
poly_model.fit(xx[:, np.newaxis], yy)
yfit = poly_model.predict(xfit[:, np.newaxis])

plt.scatter(xx, yy)
plt.plot(xfit, yfit);
```
